```python
import pandas as pd
import os
import json
import glob
from shapely.geometry  import shape, Point
from geopy.geocoders import Nominatim
import numpy as np
import meteostat as ms
import datetime as dt
```

```python
os.makedirs('./weather_data', exist_ok=True )
```

```python
start = dt.datetime(2019, 1, 1)
end = dt.datetime(2023, 12, 31)

# Create Point for chicago
chicago = ms.Point(41.85, -87.65)

data = ms.Hourly(chicago, start, end)
data = data.fetch()
data['temp_f'] = data['temp'].apply(lambda x: (x * 9/5) + 32)
data.to_csv('./weather_data/weather_hourly_data.csv')
```

```python
data_daily = ms.Daily(chicago, start, end)
data_daily = data_daily.fetch()
data_daily.head()
```

```python
data_daily = ms.Daily(chicago, start, end)
data_daily = data_daily.fetch()
data_daily[['tempmax_f', 'tempmin_f', 'tempavg_f']] = data_daily[['tmax', 'tmin', 'tavg']].apply(lambda x: (x * 9/5) + 32)
data_daily.to_csv('./weather_data/weather_daily_data.csv')
```
